```python
from math import *
import pandas as pd
import numpy as np
```

```python
inst=pd.read_csv("./instances/instance_01.csv",header=None)
inst = inst.iloc[:, 1:]
inst.head()
```

```python
list_sommet = np.arange(1,max(i for i in inst.index )+1)
list_sommet
```

```python
veh = pd.read_csv("./vehicles.csv",header=None)
veh.head()
```

```python
def c_1(R):
    mat = np.zeros[len(R),len(list_sommet)]
    for r in R : 
        j=1
        for i in r[2]:
            mat[r[j],i] = 1
        j+=1
    return mat.sum(axis=0) - np.ones(len(list_sommet)) # il faut que cette quantité soit égal à 1 pour chaque sommet, donc un vecteur de 1 de taille len(list_sommet)

```

```python
def c_2(R):
    poids = np.zeros(len(R))
    for i in range(len(R)):
        for order in R[i][2]:
            poids[i] += inst.loc[order,'order_weight' ] #le poids est en colonne 3
        poids[i]-= veh.loc[R[i][0],'max_capacity']
    return poids # il faut que cette quantité soit inférieur à 0 pour chaque route
    
```
